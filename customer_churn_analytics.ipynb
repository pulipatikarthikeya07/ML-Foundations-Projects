```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score

# 1. Create a simulated customer behavior dataset
churn_data = {
    'days_active': [30, 120, 10, 365, 45, 200, 15, 240, 300, 60],
    'monthly_spend_inr': [500, 1200, 300, 1500, 450, 1100, 600, 950, 1300, 400],
    'support_tickets_raised': [3, 1, 4, 0, 2, 1, 5, 2, 0, 3],
    'churned': [1, 0, 1, 0, 1, 0, 1, 0, 0, 1] # 1 = Left, 0 = Stayed
}

df_churn = pd.DataFrame(churn_data)

# 2. Separate features (X) and the target label (y)
X_churn = df_churn[['days_active', 'monthly_spend_inr', 'support_tickets_raised']]
y_churn = df_churn['churned']

# 3. Split into Training (70%) and Testing (30%) datasets
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_churn, y_churn, test_size=0.3, random_state=42)

# 4. Initialize and train the Decision Tree Classifier
churn_model = DecisionTreeClassifier(random_state=42)
churn_model.fit(X_train_c, y_train_c)

# 5. Run predictions on the test dataset
churn_predictions = churn_model.predict(X_test_c)

# 6. Evaluate Model performance
accuracy = accuracy_score(y_test_c, churn_predictions)

print("--- Churn Model Execution Successful ---")
print(f"Overall Model Accuracy: {accuracy * 100}%\n")
print("Detailed Performance Evaluation Matrix:")
print(classification_report(y_test_c, churn_predictions))
```
